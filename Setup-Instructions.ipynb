## **Step 5: Install Required Python Packages** 📦

### **Using `requirements.txt`**

The course includes a `requirements.txt` file with all necessary Python packages.

### **Installation Methods:**

#### **Method 1: Run Cell Below (Easiest)**
Execute the next cell to install all packages automatically.

````python
# Install required packages from requirements.txt
import sys
import subprocess
import os

print("=" * 60)
print("INSTALLING REQUIRED PACKAGES")
print("=" * 60)
print("This may take 3-5 minutes depending on your internet speed...")
print()

# Path to requirements.txt (adjust if needed)
requirements_path = os.path.join(
    os.path.dirname(os.path.abspath('__file__')),
    '..',
    '..',
    'requirements.txt'
)

# Check if requirements.txt exists
if os.path.exists(requirements_path):
    print(f"✅ Found requirements.txt at: {requirements_path}")
    print()
    
    # Install packages
    try:
        subprocess.check_call([
            sys.executable, 
            '-m', 
            'pip', 
            'install', 
            '-r', 
            requirements_path,
            '--upgrade'
        ])
        
        print()
        print("=" * 60)
        print("✅ INSTALLATION COMPLETE!")
        print("=" * 60)
        print("All required packages have been installed.")
        
    except subprocess.CalledProcessError as e:
        print()
        print("=" * 60)
        print("❌ INSTALLATION FAILED")
        print("=" * 60)
        print(f"Error: {e}")
        print()
        print("Troubleshooting:")
        print("   1. Ensure you're using the .venv Python interpreter")
        print("   2. Check your internet connection")
        print("   3. Try manual installation (see Method 2 below)")
else:
    print("❌ requirements.txt NOT FOUND")
    print(f"Expected location: {requirements_path}")
    print()
    print("Please ensure requirements.txt is in the course root directory.")
````

## **Step 6: Verify Installation** ✅

### **Comprehensive Environment Check**

Run the cell below to verify all required packages are installed correctly.

````python
# Comprehensive installation verification
import sys

print("=" * 70)
print("COMPREHENSIVE INSTALLATION VERIFICATION")
print("=" * 70)
print()

# List of required packages and their minimum versions
required_packages = {
    'numpy': '1.21.0',
    'pandas': '1.3.0',
    'matplotlib': '3.4.0',
    'seaborn': '0.11.0',
    'scikit-learn': '1.0.0',
    'flask': '2.0.0',
    'requests': '2.26.0',
    'jupyter': '1.0.0',
}

# Check each package
all_installed = True
results = []

for package, min_version in required_packages.items():
    try:
        # Import package
        if package == 'scikit-learn':
            import sklearn
            module = sklearn
            display_name = 'scikit-learn (sklearn)'
        else:
            module = __import__(package)
            display_name = package
        
        # Get version
        version = getattr(module, '__version__', 'unknown')
        
        results.append({
            'package': display_name,
            'status': '✅',
            'version': version,
            'required': f'>= {min_version}'
        })
        
    except ImportError:
        results.append({
            'package': package,
            'status': '❌',
            'version': 'NOT INSTALLED',
            'required': f'>= {min_version}'
        })
        all_installed = False

# Display results
print(f"{'Package':<30} {'Status':<8} {'Installed':<15} {'Required'}")
print("-" * 70)

for result in results:
    print(f"{result['package']:<30} {result['status']:<8} {result['version']:<15} {result['required']}")

print()
print("=" * 70)

if all_installed:
    print("🎉 SUCCESS! All required packages are installed.")
    print("=" * 70)
    print()
    print("✅ Your environment is ready for the AI Security Deep Dive course!")
    print()
    print("Next Steps:")
    print("   1. Navigate to Chapter 1 labs")
    print("   2. Open Chapter1-Labs.ipynb")
    print("   3. Start learning AI security!")
else:
    print("⚠️ INSTALLATION INCOMPLETE")
    print("=" * 70)
    print()
    print("Some packages are missing. Please:")
    print("   1. Re-run Step 5 (Install Required Packages)")
    print("   2. Check for error messages")
    print("   3. Ensure .venv is activated")
    print("   4. Contact instructor if problems persist")

print()
print("=" * 70)
print("PYTHON ENVIRONMENT SUMMARY")
print("=" * 70)
print(f"Python Version: {sys.version.split()[0]}")
print(f"Python Path: {sys.executable}")
print(f"Virtual Env: {'.venv' in sys.executable or 'venv' in sys.executable}")
print("=" * 70)
````

## **Troubleshooting Common Issues** 🔧

### **Issue 1: Python Version Too New (3.12+)**

**Problem:** Some ML libraries don't yet support Python 3.12+

**Solution:**
1. Install Python 3.11 alongside 3.12
2. Create `.venv` using Python 3.11 specifically:
   ```bash
   python3.11 -m venv .venv
   ```
3. Select Python 3.11 interpreter in VS Code

---

### **Issue 2: Virtual Environment Not Activating**

**Problem:** VS Code uses system Python instead of `.venv`

**Solution:**
1. Click Python version in **bottom-right corner** of VS Code
2. Select interpreter with `.venv` in path
3. Restart Jupyter kernel: `Kernel > Restart Kernel`

---

### **Issue 3: Package Installation Fails**

**Problem:** `pip install` errors or timeouts

**Solution:**
1. Upgrade pip:
   ```bash
   python -m pip install --upgrade pip
   ```
2. Try installing packages individually:
   ```bash
   pip install numpy pandas matplotlib seaborn scikit-learn flask requests jupyter
   ```
3. Check firewall/proxy settings

---

### **Issue 4: Jupyter Kernel Crashes**

**Problem:** Kernel dies when running cells

**Solution:**
1. Restart kernel: `Kernel > Restart Kernel`
2. Clear outputs: `Edit > Clear All Outputs`
3. Check Python memory usage (close other applications)
4. Reinstall `jupyter` and `ipykernel`:
   ```bash
   pip install --upgrade jupyter ipykernel
   ```

---

### **Issue 5: VS Code Extensions Not Working**

**Problem:** Jupyter notebooks don't render or run

**Solution:**
1. Reload VS Code: `Ctrl+Shift+P` → `Developer: Reload Window`
2. Reinstall extensions:
   - Uninstall Python, Jupyter, Pylance extensions
   - Restart VS Code
   - Reinstall extensions
3. Update VS Code to latest version
